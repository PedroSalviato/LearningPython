```python
import pandas as pd
import numpy as np
```

```python
## Nesse caso não contém valores nulos nas colunas mas por precaução vamos eliminar os nulos
df_value = df['Value'].fillna('0', inplace = True) 
df_wage = df['Wage'].fillna('0', inplace = True)
```

```python
# pegando o símbolo de euro e armazenando em uma varíavel

simbol_euro = df['Value'].values[0][0] #pegando o símbolo de euro e armazenando na varíavel
print(simbol_euro)
```

```python
# Função para tratar os campos
def func_replace(value):
    valor_campo = value.replace(simbol_euro, '')
    if 'M' in valor_campo:
        valor_campo = valor_campo.replace('M', '') 
        return float(valor_campo)*1000000
    elif 'K' in valor_campo:
        valor_campo = valor_campo.replace('K', '')
        return float(valor_campo)*1000
```

```python
 # Aplicando função
df['Value'] = df['Value'].apply(func_replace)
df['Wage'] = df['Wage'].apply(func_replace)
```

```python
# Perceba que os campos foram tratados conforme a função
df.head()
```

```python
# Tratando a coluna Release Clause com loops
release_clause = []

for valor_campo in df['Release Clause'].values:
    valor_sem_euro = valor_campo.replace(simbol_euro, '')
    if 'M' in valor_sem_euro:
        valor_sem_euro = float(valor_sem_euro.replace('M', '')) * 1000000
    elif 'K' in valor_sem_euro:
        valor_sem_euro = float(valor_sem_euro.replace('K', '')) *1000
    release_clause.append(valor_sem_euro)
df['Release Clause1'] = release_clause
    
```
