### 2. Filtrando dados

```python
df.loc[0:5, 'Name']
```

```python
# utilizandoo o LOC e especificando algumas colunas
df.loc[0:6, ['Name', 'Potential', 'Age']]
```

```python
# Utilizando o ILOC e selecionando o intervalo de linhas e colunas

df.iloc[0:10, 0:7]
```

```python
# Filtrando com base em uma expressão

df.loc[df['Age'] <= 18].head(5)
```

```python
# Filtrando com o filter
df.filter(['Name', 'Age', 'Nationality'])
```

```python
# filtrando com base na média da idade dos jogadores

media_age = df['Age'].mean()
print(media_age)
# Usando o LOC
df.loc[df['Age'] <= media_age].head(5)
```

```python
# Usando o QUERY
df.query('Age <= @media_age').head(5) ## utilizo o @ pois a variavel vem de fora do dataset
```

```python
# Filtragem contendo uma lista de valores
df_nacional =  df[df['Nationality'].isin(['Brazil'])]
df_nacional.head()
```

```python
# Filtragem comparando valores
df_comp = df[df['Potential'] == df['Overall']]
df_comp.head()
```

```python
# Filtrando colunas númericas

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

df.numerics = df.select_dtypes(exclude = numerics) # posso utilizar o include aqui também
df.numerics.head()
```
